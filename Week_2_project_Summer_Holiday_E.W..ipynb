Load the CSV file into a dataframe, skip any extra initial spaces, set the 'Date' as index:

## Finding a summer break

According to meteorologists, summer extends for the whole months of June, July, and August in the northern hemisphere and the whole months of December, January, and February in the southern hemisphere. So as I'm in the northern hemisphere I'm going to create a dataframe that holds just those months using the `datetime` index, like this:

```python
summer = beijing.loc['2014-6-1':'2014-8-31' ]
```

Summer in Beijing is typically hot, According to [_Wikipedia_](https://en.wikipedia.org/wiki/Beijing#Climate),
 Beijing has a monsoon-influenced humid continental climate, characterized by higher humidity in the summers due to the East Asian monsoon. Precipitation averages around 570 mm (22 in) annually, with close to three-fourths of that total falling from June to August. 

So I am going to focus on three aspects: Temperature, Humidity, Precipitation

```python
focus = ['Max TemperatureC', 'Mean TemperatureC', 'Min TemperatureC', 'Max Humidity', 'Mean Humidity', 'Min Humidity', 'Precipitationmm']
summer_focus = summer[focus]
summer_focus.describe()
summer_focus.head(14)
```

It is better to see these data in graphs, so lets tell Jupyter to display any graph created inside this notebook:

```python
%matplotlib inline
```

Mean Temperature, Mean Humidity and Precipitation during the summer period

```python
summer_focus[['Mean TemperatureC', 'Mean Humidity', 'Precipitationmm']].plot(grid=True, figsize=(10,5))
```

Well looking at the graph, while the mean humidity fluctuated in a wide range (from 34% to 86%) during the whole summer period, the second half of August looks good for relative low mean temperatures and precipitation.

Let's check the humidity factor in more details:<br/>
[_Human Comfort_](https://en.wikipedia.org/wiki/Humidity#Human_comfort): Humans are sensitive to humid air because the human body uses evaporative cooling as the primary mechanism to regulate temperature. Under humid conditions, the rate at which perspiration evaporates on the skin is lower than it would be under arid conditions. Because humans perceive the rate of heat transfer from the body rather than temperature itself, we feel warmer when the relative humidity is high than when it is low.

Taking temperature and humidity into account, there is a comfort zone for human.<br/>
According to [_ISO7730_](https://www.azosensors.com/Article.aspx?ArticleID=487), for the summer period, the comfort zone is defined as:<br/>
Temperature : 23 ~ 26 Celsius degrees<br/>
Relative Humidity: 30% ~ 70%

Here I use the `'Mean TemperatureC'` and `'Mean Humidity'` to identify the confort days

```python
comfort_t = (summer_focus['Mean TemperatureC'] >= 23) & (summer_focus['Mean TemperatureC'] <= 26)
comfort_h = (summer_focus['Mean Humidity'] >=30) & (summer_focus['Mean Humidity'] <= 70)
comfort_days = summer_focus[comfort_t & comfort_h]
comfort_days.describe()
```

There were 27 comfort days out of the 92-day summer time.

Check how these comfort days distributed in a weekly basis:

```python
comfort_days['Precipitationmm'].resample('W').count().plot(kind='bar', grid=True, figsize=(10,3))
```

It is clear that the comfort days are mainly in June and August


Now let's plot the Weekly Sunny Days for the summer:

```python
sunny_days = summer_focus[summer_focus['Precipitationmm'] == 0]
sunny_days['Precipitationmm'].resample('W').count().plot(kind='bar', grid=True, figsize=(10,5))
```

Taking the total weekly sunny days into consideration,  the two weeks from 2014-Aug-10 to 2014-Aug-23 (with 10 sunny days and 7 comfort days) seems to be the best time for the summer vocation.

Let's check this period's temperature and precipitation again to comfirm the above results:

```python
holiday = summer_focus.loc['2014-8-10':'2014-8-23']
holiday.describe()
```

```python
holiday[['Mean TemperatureC', 'Precipitationmm']].plot(grid=True, figsize=(10,5))
```

Yes, it looks pretty good, the mean temperature is between 20 and 28 degrees and there is a 7-day 0 precipitation period in between.
